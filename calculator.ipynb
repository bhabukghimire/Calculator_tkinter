```python
from tkinter import *

root = Tk()
root.title("Simple Calculator")
root.config(bg="#17202a")

def callback():
    print("Call back function has been called.")
    
#row0 for display
display = Entry(root, relief="raised", font=("Cursive",30) , fg = 'black')
display.insert(0,"0")
display.grid(row=0,column=0,columnspan=5,)


#operations
def clear():
    display.delete(0,END)
    display.insert(0,"0")
    
def back():
    val=display.get()s
    display.delete(len(val)-1,END)
    
def put_value(val):
    if(display.get()== "0"):
        display.delete(0,END)
        display.insert(END,str(val))
    else:
        display.insert(END, str(val))
        
def set_op(op):
    num=display.get()
    global operand
    global number1
    number1=float(num)
    operand = str(op)
    display.delete(0,END)
    
def equal():
    global operand
    global number1
    number2 = float(display.get())
    display.delete(0,END)
    if operand == "+":
        result = number1 + number2
    if operand == "-":
        result = number1 - number2
    if operand == "X":
        result = number1 * number2
    if  operand == "/":
        result = number1 / number2
   
        
    display.insert(0,str(result))
        
        
    
    


#row1
button_plus=Button(root, text="+", command=lambda:set_op("+"),bg="lightblue",activebackground="pink").grid(row=1,column=0, ipadx=24, ipady=10, padx=1, pady=1)
button_minus=Button(root, text="-", command=lambda:set_op("-"),bg="lightblue",activebackground="pink").grid(row=1,column=1,ipadx=24, ipady=10, padx=1, pady=1)
button_clear=Button(root, text="C", command=clear ,bg="red",activebackground="yellow").grid(row=1,column=2,ipadx=24, ipady=10, padx=1, pady=1)
button_back=Button(root, text="Del", command=back ,bg="red",activebackground="yellow").grid(row=1,column=3,ipadx=24, ipady=10, padx=1, pady=1)

#row2
button7=Button(root, text="7", command=lambda: put_value(7) ,bg="#566573",activebackground="#eaecee").grid(row=2,column=0,ipadx=24, ipady=10, padx=1, pady=1)
button8=Button(root, text="8", command=lambda: put_value(8),bg="#566573",activebackground="#eaecee").grid(row=2,column=1,ipadx=24, ipady=10, padx=1, pady=1)
button9=Button(root, text="9", command=lambda: put_value(9),bg="#566573",activebackground="#eaecee").grid(row=2,column=2,ipadx=24, ipady=10, padx=1, pady=1)
button_multiply=Button(root, text="X", command=lambda:set_op("X"),bg="lightblue",activebackground="pink").grid(row=2,column=3,ipadx=24, ipady=10, padx=1, pady=1)

#row3
button4=Button(root, text="4", command=lambda: put_value(4),bg="#566573",activebackground="#eaecee").grid(row=3,column=0,ipadx=24, ipady=10, padx=1, pady=1)
button5=Button(root, text="5", command=lambda: put_value(5),bg="#566573",activebackground="#eaecee").grid(row=3,column=1,ipadx=24, ipady=10, padx=1, pady=1)
button6=Button(root, text="6", command=lambda: put_value(6),bg="#566573",activebackground="#eaecee").grid(row=3,column=2,ipadx=24, ipady=10, padx=1, pady=1)
button_divide=Button(root, text="/", command=lambda:set_op("/"),bg="lightblue",activebackground="pink").grid(row=3,column=3,ipadx=24, ipady=10, padx=1, pady=1)

#row4
button1=Button(root, text="1", command=lambda: put_value(1),bg="#566573",activebackground="#eaecee").grid(row=4,column=0,ipadx=24, ipady=10, padx=1, pady=1)
button2=Button(root, text="2", command=lambda: put_value(2),bg="#566573",activebackground="#eaecee").grid(row=4,column=1,ipadx=24, ipady=10, padx=1, pady=1)
button3=Button(root, text="3", command=lambda: put_value(3),bg="#566573",activebackground="#eaecee").grid(row=4,column=2,ipadx=24, ipady=10, padx=1, pady=1)
button_equal=Button(root, text="=", command=equal,bg="gold",activebackground="#6c3483").grid(row=4,column=3, rowspan=2,ipadx=24, ipady=37, padx=1, pady=1)

#row5
button0 = Button(root,text="0", command=lambda: put_value(0),bg="#566573",activebackground="#eaecee").grid(row=5,column=0,columnspan=2,ipadx=62, ipady=10, padx=1, pady=1)
button_point = Button(root,text=".",command= lambda: put_value("."),bg="lightblue",activebackground="pink").grid(row=5,column=2,ipadx=24, ipady=10, padx=1, pady=1)



root.mainloop()
```
